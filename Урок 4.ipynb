Равномерное распределение: $$M(X) = \frac{a + b}{2}$$ $$D(X) = \frac{(b - a)^2}{12}$$

```python
# 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
# Найдите ее среднее значение и дисперсию.

print(f'Среднее значение: {(200 + 800) / 2}')
print(f'Дисперсия: {(800 -200) ** 2 / 12}')
```

```python
# 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
# Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
# Если да, найдите ее.

D = 0.2  # дисперсия
a = 0.5  # левая граница

# Из формулы дисперсии считаем правую границу интервала b

b = np.sqrt(D * 12) + a
print(f'Правая граница: {b}')
print(f'Среднее значение: {(a + b) / 2}')   
```

Функция плотности нормального распределения: $$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}}$$

```python
# 3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения
# f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-(x+2)**2) / 32).
# Найдите:
# а). M(X)
# б). D(X)
# в). std(X) (среднее квадратичное отклонение)

# Решение:
# (x-a) = (x+2) => M(X)=-2
# 2*σ2 = 2*D(X) = 32 => D(X)=16
# std(X) = σ = sqrt(D(X)) = 4
```

### Правило трех сигм
На отрезке от $-\sigma$ до $+\sigma$ расположено около 68 % наблюдений, 

от $-2\sigma$ до $+2\sigma$ — 95.4 %,

и от $-3\sigma$ до $+3\sigma$ — 99.72 %.

```python
# 4. Рост взрослого населения города X имеет нормальное распределение.
# Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
# Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
# а). больше 182 см
# б). больше 190 см
# в). от 166 см до 190 см
# г). от 166 см до 182 см
# д). от 158 см до 190 см
# е). не выше 150 см или не ниже 190 см
# ё). не выше 150 см или не ниже 198 см
# ж). ниже 166 см.

# Решение: т.к все заданные точки располагаются от среднего значения на расстояниях кратных СКО,
#          тогда воспользуемся правилом трех сигм

print(f'а) больше 182 см: {1 - 0.5 - 0.68/2}')  # площадь правее +std(X)
print(f'б) больше 190 см: {1 - 0.5 - 0.954/2}')  # площадь правее +2*std(X)
print(f'в) от 166 см до 190 см: {0.68/2 + 0.954/2}')  # площадь от -std(X) до +2*std(X)
print(f'г) от 166 см до 182 см: {0.68}')  # площадь от -std(X) до +std(X)
print(f'д) от 158 см до 190 см: {0.954}')  # площадь от -2*std(X) до +2*std(X)
print(f'е) не выше 150 см или не ниже 190 см: {1 - (0.9972/2 + 0.954/2)}')  # площадь до -3*std(X) и правее +2*std(X)
print(f'ё) не выше 150 см или не ниже 198 см: {1 - 0.9972}')  # площадь до -3*std(X) и правее +3*std(X)
print(f'ж) ниже 166 см: {0.5 - 0.68/2}')  # площадь левее -std(X)
```

```python
# 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
# математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

# Решение: сигма = корень из D(X) = 5 см.
print(f'{(190 - 178) / 5} сигмы')
```
