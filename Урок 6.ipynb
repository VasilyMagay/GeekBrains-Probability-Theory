```python
# 1) Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):
# zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
# ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].
# Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy
# Полученные значения должны быть равны.
# Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков,
# а затем с использованием функций из библиотек numpy и pandas.

import numpy as np

def my_mean(arr):  # среднее арифметическое
    return arr.sum() / len(arr)

def my_std(arr):  # среднеквадратичное отклонение
    mean = my_mean(arr)
    return np.sqrt(((arr - mean)**2).sum() / len(arr))


zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

$$cov_{XY} = M[(X - M(X))(Y - M(Y))] = M(XY) - M(X)M(Y)=\overline{X \cdot Y} - \overline{X} \cdot \overline{Y}$$

```python
c = my_mean(zp * ks) - my_mean(zp) * my_mean(ks)
cn = np.cov(zp, ks, ddof=0)
print(f'Коварианция = {c}, проверка по numpy = {cn[0,1]}')
```

$$r_{XY} = \frac{cov_{XY}}{\sigma_{X}\sigma_{Y}}$$

```python
r = c/(my_std(zp) * my_std(ks))  # коэффициент корреляции Пирсона
rn = np.corrcoef(zp, ks)
print(f'Коэффициент корреляции Пирсона = {r}, проверка по numpy = {rn[0,1]}')
```

```python
# 2) Измерены значения IQ выборки студентов,
# обучающихся в местных технических вузах:
# 131, 125, 115, 122, 131, 115, 107, 99, 125, 111.
# Известно, что в генеральной совокупности IQ распределен нормально.
# Найдите доверительный интервал для математического ожидания с надежностью 0.95.

# Решение: сначала рассчитаем сигму и среднее для опытных данных
# Будем использовать t-критерий, т.к. сигма ген.совокупности неизвестна

x = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
D = np.var(x, ddof=1)  # оценка дисперсии, несмещенная
M = np.mean(x)
print(f'M={M} D={D}')

a = 0.05  # 5%
n = 10
t = 2.262  # значение для объема выборки 10 и 95% дов.интервала, но для a/2

print(f'Доверительный интервал: [{M - t * np.sqrt(D/n)}; {M + t * np.sqrt(D/n)}]')
```

```python
# 3) Известно, что рост футболистов в сборной распределен нормально
# с дисперсией генеральной совокупности, равной 25 кв.см. Объем выборки равен 27,
# среднее выборочное составляет 174.2. Найдите доверительный интервал для математического
# ожидания с надежностью 0.95.

# Решение: сигма ген. совокупности известна (sqrt(25)=5 см.), используем z-критерий в формуле доверительного интервала
#  M ± z(a/2) * std / sqrt(n)

from scipy.stats import norm

M = 174.2  # выборочная средняя
a = 0.05  # уровень значимости 5 %
n = 27  # объем выборки
D = 25  # дисперсия ген.совокупности
sigma = np.sqrt(D)  # среднеквадратичное отклонение

print(f'Доверительный интервал: [{M + norm.ppf(a/2) * sigma / np.sqrt(n)}; {M - norm.ppf(a/2) * sigma / np.sqrt(n)}]')
```
